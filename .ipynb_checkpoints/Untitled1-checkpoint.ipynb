```python
import matplotlib.pyplot as plt


student_attendance = {'day1':33, 'day2':34,'day3':29,'day4':31,'day5':28,'day6':26,'day7':30}



xline = student_attendance.keys()
yline = student_attendance.values()

plt.plot(xline, yline, linewidth=5, c="red")
plt.figure()
plt.plot(xline, yline, linewidth=5)


 # Set chart title and label axes. 
plt.title("student_attendance", fontsize=24)
plt.xlabel("Days", fontsize=14)
plt.ylabel("number of students", fontsize=14)
# Set size of tick labels.

plt.tick_params(axis='both', labelsize=10)



```
